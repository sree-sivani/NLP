###LEMMATIZATION

This technique is similar to stemming.
The output we will get here is called lemma.
Lemma is a root word rather than root stem.
Using this we will get the root word which is meaningful unlike as in Stemming.

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going", pos = 'v')
```

```python
lemmatizer.lemmatize("fairly", pos = 'n')
```

```python
lemmatizer.lemmatize("lovingly", pos = 'a')
```
