###LEMMATIZATION

This technique is similar to stemming.
The output we will get here is called lemma.
Lemma is a root word rather than root stem

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```
