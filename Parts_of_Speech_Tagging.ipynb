```python
paragraph = """In *All the Bright Places* by Jennifer Niven, Theodore Finch is a complex character grappling with mental illness and deep emotional struggles. He is charismatic and unpredictable, often labeled as the school "freak," but beneath his wild exterior, Finch hides profound feelings of isolation and pain. His fascination with death and frequent suicidal thoughts contrast sharply with moments of intense creativity and connection, particularly in his relationship with Violet Markey. Throughout the novel, Finch is on a journey to find meaning and escape his inner turmoil, symbolizing both the beauty and fragility of life. His character highlights the often-hidden struggles of mental health and the need for understanding and support.Finch is a character who oscillates between highs and lows, constantly searching for ways to stay awake in a world where he feels he doesn’t quite belong. His relationship with Violet becomes a lifeline, offering moments of joy and purpose as they explore Indiana’s hidden wonders together. However, despite his deep affection for Violet and his efforts to stay grounded, Finch is haunted by his own inner demons. His mental health deteriorates, and his inability to find lasting peace highlights the tragic nature of his character. Finch’s journey is a poignant reminder of the complexities of mental illness and the profound impact it can have on even the most vibrant lives."""
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[word for word in words if word not in set(stopwords.words('english'))]
    #sentences[i]=' '.join(words)# converting all the list of words into sentences
    pos_tag=nltk.pos_tag(words)
    print(pos_tag)
```
