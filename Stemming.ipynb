###STEMMING

Stemming is the process of reducing a word to its 'word stem' that affixes to suffixes and prefixes or to the roots of words known as lemma.

It is imp in NLU -> Natural Language Understanding


```python
words = ["Eating", "eats", "eaten", "writing", "writes", "programming", "program", "history", "finally", "finalized", "apple"]
```

# PorterStemmer Class

```python
#PorterStemmer
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in words:
  print(word + " ---> " + stemming.stem(word))
```

```python
stemming.stem("Congratulations")
stemming.stem("sitting")
```

# RegexpStemmer Class


```python
#RegexpStemmer class
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
reg_stemmer
reg_stemmer.stem("Congratulations")
```

### SNOWBALL STEMMER


```python
#snowball stemmer - better than porter stemmer
#gives better accuracy than porter -> better form of word
```

```python
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer('english')
```

```python
for word in words:
  print(word + " --> " +  snowball_stemmer.stem(word))
```

DIFFERENCE

```python
#difference -> useful in text processing
stemming.stem("fairly"), stemming.stem("sportingly")
```

```python
snowball_stemmer.stem("fairly"), snowball_stemmer.stem("sportingly")
```
